```python
import seaborn as sns
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
def plot_word_frequency(words,top_n=10):
    word_freq=FreqDist(words)
    labels=[element[0] for element in word_freq.most_common(top_n) ]
    count=[element[1] for element in word_freq.most_common(top_n)]
    plot=sns.barplot(x=labels,y=count)
    return plot
```

```python
def remove_stop_words(words):
    no_stops = [word for word in words if word.lower() not in stopwords.words("english")]
    return no_stops
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize_words=word_tokenize(document)
```

```python
word_tokenize_words
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize_words=sent_tokenize(document)
sent_tokenize_words
```

```python
from nltk.tokenize import TweetTokenizer
tweet_tokenizer=TweetTokenizer()
tweet_tokenizer_words=tweet_tokenizer.tokenize(document)
tweet_tokenizer_words
```

```python
from nltk.tokenize import regexp_tokenize
pattern = "[\w]+"
regexp_tokenize_words=regexp_tokenize(document,pattern)
regexp_tokenize_words
```

```python
no_stops_words=remove_stop_words(regexp_tokenize_words)
```

```python
plot_word_frequency(no_stops_words,15)
```
